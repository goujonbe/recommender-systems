```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('rmse_trend.csv')
df.head()
```

```python
df.plot(x='size', y='RMSE', style='o')
plt.ylim(0)
```

```python
poly = PolynomialFeatures(degree=2)
X = poly.fit_transform(np.array(df['size']).reshape(-1,1))
y = df['RMSE']
model = LinearRegression()
model.fit(X, y)
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(20,10))
plt.scatter(df['size'], df['RMSE'], marker='.', linewidth=5, label='valeurs réelles')
plt.scatter(df['size'], model.predict(X), marker='.', color='firebrick', linewidth=5, label='prédiction')
plt.ylim(0)
plt.legend(loc='upper right')
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 22})
plt.scatter(df['size'], df['RMSE'], marker='.', linewidth=5, label='valeurs réelles')
plt.hlines(y=1, xmin=df['size'].min(), xmax=df['size'].max(), linestyle='solid', color='green', label='y=1')
plt.hlines(y=0.95, xmin=df['size'].min(), xmax=df['size'].max(), linestyle='solid', color='limegreen', label='y=0.95')
plt.hlines(y=0.9, xmin=df['size'].min(), xmax=df['size'].max(), linestyle='solid', color='lime', label='y=0.9')
plt.ylim(0)
plt.grid(True, linestyle=':', linewidth=2)
plt.legend(loc='lower right')
```
